```python
import pandas as pd
import matplotlib.pyplot as plt

print("Libraries loaded successfully!")
```

```python
# load CSV's
df_basic = pd.read_csv("CSVresults/BasicStrategy_exportedCSV.csv", header=None)
df_hilo = pd.read_csv("CSVresults/HiLoStrategy_exportedCSV.csv", header=None)

# Clean the extra comma
df_basic = df_basic.dropna(axis=1, how='all')
df_hilo = df_hilo.dropna(axis=1, how='all')

print("Data loaded!")
```

```python
# mean calc
mean_basic = df_basic.mean(axis=0)
mean_hilo = df_hilo.mean(axis=0)

# drawing
plt.figure(figsize=(10, 6))

plt.plot(mean_basic, label="Basic Strategy", color='red')
plt.plot(mean_hilo, label="HiLo Strategy (Card Counting)", color='green')

plt.title("MonteCarlo simulation: Basic vs HiLo Strategy (100,000 Hands played)")

plt.xlabel("No. of Hands Played")
plt.ylabel("Average Net Profit")

plt.axhline(0, color='black', linestyle='--', label="Break Even")

plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

strategies = {
    "HiLo (Card Counting)": {"file": "HiLoStrategy_exportedCSV.csv", "color": "green"},
    "Random Play": {"file": "RandomStrategy_exportedCSV.csv", "color": "gray"},
    "Basic Strategy": {"file": "BasicStrategy_exportedCSV.csv", "color": "blue"},
    "Threshold (stand on 17)": {"file": "ThresholdStrategy_exportedCSV.csv", "color": "orange"},
    "Only Stand": {"file": "OnlyStandStrategy_exportedCSV.csv", "color": "purple"},
    "Only Hit": {"file": "OnlyHitStrategy_exportedCSV.csv", "color": "red"}
}

plt.figure(figsize=(12, 8))

for name, info in strategies.items():
    try:
        df = pd.read_csv(f"CSVresults/{info['file']}", header=None)
        df = df.dropna(axis=1, how='all')
        
        mean_vals = df.mean(axis=0)
        
        plt.plot(mean_vals, label=name, color=info['color'], linewidth=2)
        print(f"Successfully plotted {name}")
    except FileNotFoundError:
        print(f"Skipping {name} - File not found!")

plt.title("MonteCarlo simulation: Blackjack Strategies Compared (100,000 Hands)", fontsize=16)
plt.xlabel("No. of Hands Played", fontsize=12)
plt.ylabel("Average Net Profit", fontsize=12)
plt.axhline(0, color='black', linestyle='--', linewidth=2, label="Break Even")

plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

filename = "CSVresults/HiLoStrategy_exportedCSV.csv" 
strategy_name = "HiLo Strategy (Card Counting)"

df = pd.read_csv(filename, header=None)
df = df.dropna(axis=1, how='all')

plt.figure(figsize=(12, 8))

# alpha=0.1 for 90% transparency
for i in range(len(df)):
    plt.plot(df.iloc[i], color='green', alpha=0.1)

mean_vals = df.mean(axis=0)
plt.plot(mean_vals, color='black', linewidth=3, label="Overall Average (Expected Value)")

plt.title(f"Monte Carlo Variance: 1000 Players using {strategy_name}", fontsize=16)
plt.xlabel("Number of Hands Played", fontsize=12)
plt.ylabel("Net Profit", fontsize=12)
plt.axhline(0, color='red', linestyle='--', linewidth=2, label="Break Even ($0)")

plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()

plt.show()
```
